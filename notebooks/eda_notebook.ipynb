```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Corrected file path (relative to the root project directory)
file_path = "notebooks/../data/processed/user_interactions_cleaned.csv"

# Load the dataset
df = pd.read_csv(file_path)

# Convert timestamp to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Display basic statistics
display(df.describe())

# Plot rating distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["rating"], bins=5, kde=True)
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Distribution of Ratings")
plt.show()

# Analyze most popular items
popular_items = df["item_id"].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=popular_items.index, y=popular_items.values)
plt.xlabel("Item ID")
plt.ylabel("Number of Interactions")
plt.title("Top 10 Most Interacted Items")
plt.show()

# Analyze most active users
active_users = df["user_id"].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=active_users.index, y=active_users.values)
plt.xlabel("User ID")
plt.ylabel("Number of Interactions")
plt.title("Top 10 Most Active Users")
plt.show()

print("✅ EDA Completed Successfully!")
```
